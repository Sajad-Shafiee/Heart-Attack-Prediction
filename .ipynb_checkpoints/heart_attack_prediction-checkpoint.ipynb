# پروژه Classification: پیش‌بینی شانس حمله قلبی (Heart Attack Risk)

هدف: با استفاده از داده‌های پزشکی، پیش‌بینی کنیم آیا بیمار:
- **شانس کم** برای حمله قلبی دارد (کلاس 0)
- **شانس زیاد** برای حمله قلبی دارد (کلاس 1)

نکته‌ی کلیدی حوزه سلامت:
- **FN (False Negative)** بسیار خطرناک است؛ یعنی بیمار پرخطر را کم‌خطر تشخیص داده‌ایم.
- بنابراین به جای F1 از **F2-score** استفاده می‌کنیم (وزن بیشتر به Recall).

در این نوت‌بوک:
1) بارگذاری و تحلیل اولیه داده  
2) پیش‌پردازش + رسم نمودارهای کارآمد (توزیع‌ها و همبستگی)  
3) آموزش و بهینه‌سازی حداقل ۳ مدل (SVM, KNN, DecisionTree)  
4) گزارش نهایی هر مدل: classification_report + F2-score + Confusion Matrix  
5) مقایسه نهایی بر اساس F2 و FN


## 1) ایمپورت کتابخانه‌ها و تنظیمات پایه

- از `sklearn` برای مدل‌ها و ارزیابی استفاده می‌کنیم.
- برای تکرارپذیری نتایج، یک `RANDOM_SEED` ثابت تعریف می‌کنیم.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    fbeta_score,
    make_scorer,
    accuracy_score
)

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

RANDOM_SEED = 19
np.random.seed(RANDOM_SEED)
```

## 2) بارگذاری داده و بررسی اولیه

در این مرحله:
- فایل داده را می‌خوانیم (با پوشش دو حالت: کنار نوت‌بوک یا مسیر `/mnt/data/`).
- ابعاد داده و چند ردیف اول را می‌بینیم.
- مقدارهای گمشده (Missing) را بررسی می‌کنیم.
- توزیع کلاس هدف را بررسی می‌کنیم تا ببینیم داده نامتوازن است یا نه.

نکته: در این دیتاست ستون هدف `output` است.


```python
import os

possible_paths = ["heart.csv", "/mnt/data/heart.csv"]

data_path = None
for p in possible_paths:
    if os.path.exists(p):
        data_path = p
        break

if data_path is None:
    raise FileNotFoundError("heart.csv not found. Put it next to notebook or in /mnt/data/")

df = pd.read_csv(data_path)
print("Loaded from:", data_path)

print("Shape:", df.shape)
display(df.head())

print("\nMissing values per column:")
display(df.isna().sum())

print("\nTarget distribution (output):")
display(df["output"].value_counts())
display(df["output"].value_counts(normalize=True))
```

## 3) معنی ستون‌ها (برای خوانایی گزارش)

این بخش صرفاً برای مستندسازی است تا گزارش پروژه واضح و قابل دفاع باشد.


```python
feature_description = {
    "age": "سن",
    "sex": "جنسیت (1=مرد، 0=زن)",
    "cp": "نوع درد قفسه سینه (کدگذاری شده)",
    "trtbps": "فشار خون در حالت استراحت",
    "chol": "کلسترول",
    "fbs": "قند خون ناشتا بالا (1=دارد،0=ندارد)",
    "restecg": "نتیجه نوار قلب در حالت استراحت (کدگذاری شده)",
    "thalachh": "حداکثر ضربان قلب",
    "exng": "درد با فعالیت (1=بله،0=خیر)",
    "oldpeak": "افت ST",
    "slp": "شیب ST (کدگذاری شده)",
    "caa": "تعداد رگ‌های اصلی (کدگذاری شده)",
    "thall": "thal (کدگذاری شده)",
    "output": "هدف (0=شانس کم، 1=شانس زیاد)"
}
pd.Series(feature_description)
```

## 4) نمودارهای اولیه

در این مرحله:
- توزیع کلاس هدف را به صورت نمودار ستونی مشاهده می‌کنیم.
- چند ویژگی عددی مهم را به شکل هیستوگرام بررسی می‌کنیم تا:
  - محدوده‌ها را ببینیم
  - وجود outlier را حدس بزنیم
  - به ضرورت Scaling برای بعضی مدل‌ها (مثل SVM و KNN) برسیم


```python
# توزیع کلاس‌ها
df["output"].value_counts().plot(kind="bar")
plt.title("Target distribution (0=low risk, 1=high risk)")
plt.xlabel("output")
plt.ylabel("count")
plt.show()

# توزیع چند ویژگی عددی مهم
for col in ["age", "trtbps", "chol", "thalachh", "oldpeak"]:
    df[col].hist(bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

## 5) بررسی همبستگی (Correlation) برای ویژگی‌های عددی

هدف:
- مشاهده رابطه‌ی خطی تقریبی بین ویژگی‌های عددی و خروجی.
- کمک به تحلیل اولیه و انتخاب مسیر پیش‌پردازش/مدل.

نکته:
- Correlation برای ویژگی‌های صرفاً دسته‌ای خیلی معنی‌دار نیست.
- اینجا فقط عددی‌ها را بررسی می‌کنیم.


```python
numeric_cols = ["age", "trtbps", "chol", "thalachh", "oldpeak"]
corr = df[numeric_cols + ["output"]].corr()

plt.figure(figsize=(7, 6))
plt.imshow(corr, aspect="auto")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation (numeric features + output)")
plt.colorbar()
plt.tight_layout()
plt.show()

display(corr)
```

## 6) آماده‌سازی X و y + تعیین ستون‌های دسته‌ای/عددی

- `y` همان ستون هدف (output) است.
- `X` شامل بقیه ویژگی‌هاست.

در این دیتاست ستون‌هایی که ماهیت دسته‌ای/کدگذاری‌شده دارند:
`sex, cp, fbs, restecg, exng, slp, caa, thall`

برای سادگی و عملکرد مناسب:
- در این پروژه همه ستون‌ها را عددی فرض می‌کنیم،
- سپس روی همه ستون‌ها Impute و StandardScaler می‌زنیم.
این کار مخصوصاً برای **SVM و KNN** بسیار مهم است.


```python
X = df.drop(columns=["output"])
y = df["output"]

cat_cols = ["sex", "cp", "fbs", "restecg", "exng", "slp", "caa", "thall"]
num_cols = [c for c in X.columns if c not in cat_cols]

print("Categorical-like columns:", cat_cols)
print("Numeric columns:", num_cols)
print("\nAll feature columns:", list(X.columns))
```

## 7) ساخت پیش‌پردازش استاندارد

اقدامات:
- Imputer (median) برای اطمینان از نبود مقدار گمشده
- StandardScaler برای هم‌مقیاس کردن ویژگی‌ها

این پیش‌پردازش را داخل Pipeline می‌گذاریم تا:
- نشت اطلاعات (data leakage) اتفاق نیفتد
- همه مراحل در CV و GridSearch درست اجرا شوند


```python
preprocess_all_numeric = ColumnTransformer(
    transformers=[
        ("all", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]), X.columns)
    ],
    remainder="drop"
)
```

## 8) تقسیم داده به Train/Test

- از `stratify=y` استفاده می‌کنیم تا نسبت کلاس‌ها در train و test تقریباً ثابت بماند.
- test_size را 0.2 می‌گذاریم.


```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=RANDOM_SEED,
    stratify=y
)

print("Train size:", X_train.shape, "Test size:", X_test.shape)
print("\nTrain target distribution:")
display(y_train.value_counts(normalize=True))
print("\nTest target distribution:")
display(y_test.value_counts(normalize=True))
```

## 9) تعریف تابع ارزیابی استاندارد مدل‌ها

برای هر مدل (بعد از انتخاب بهترین تنظیمات):
- Accuracy را حساب می‌کنیم.
- F2-score را با pos_label=1 حساب می‌کنیم (کلاس پرخطر).
- classification_report می‌گیریم.
- Confusion Matrix رسم می‌کنیم با لیبل‌های درست.

تحلیل پزشکی:
- FN مهم‌ترین خطاست: بیمار پرخطر را کم‌خطر تشخیص داده‌ایم.
- در نتیجه تصمیم‌گیری نهایی را بر اساس **F2 و FN** انجام می‌دهیم.


```python
def evaluate_model(model, X_test, y_test, labels=("شانس کم (0)", "شانس زیاد (1)")):
    """
    گزارش استاندارد برای پروژه سلامت:
    - Accuracy
    - F2 (pos_label=1)
    - classification_report
    - confusion matrix + FN/FP
    - تحلیل پزشکی عددی (Recall, FNR, Specificity, FPR)
    """
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    f2 = fbeta_score(y_test, y_pred, beta=2, pos_label=1)

    print("Accuracy:", round(acc, 4))
    print("F2-score (positive=1):", round(f2, 4))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=list(labels)))

    cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()

    # نرخ‌های مهم برای تحلیل پزشکی
    recall_pos = tp / (tp + fn) if (tp + fn) else 0.0       # حساسیت برای کلاس 1
    fnr = fn / (fn + tp) if (fn + tp) else 0.0              # نرخ از دست دادن بیماران پرخطر
    specificity = tn / (tn + fp) if (tn + fp) else 0.0      # ویژگی برای کلاس 0
    fpr = fp / (fp + tn) if (fp + tn) else 0.0              # نرخ هشدار اشتباه

    print("Confusion Matrix [ [TN, FP], [FN, TP] ]:")
    print(cm)
    print(f"TN={tn}, FP={fp}, FN={fn}, TP={tp}")
    print("جمع False ها (FP+FN) =", fp + fn)

    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=list(labels))
    disp.plot(values_format="d")
    plt.title("Confusion Matrix")
    plt.show()

    print("\nتحلیل پزشکی (عددی و دقیق):")
    print(f"- Recall(کلاس پرخطر=1) = {recall_pos:.3f}  |  FNR = {fnr:.3f}")
    print(f"- Specificity(کلاس کم‌خطر=0) = {specificity:.3f}  |  FPR = {fpr:.3f}")
    print(f"- مهم‌ترین خطا در سلامت: FN={fn} (بیمار پرخطر را کم‌خطر تشخیص داده‌ایم).")
    if fn == 0:
        print("- نتیجه: از نظر بالینی، این مدل هیچ بیمار پرخطری را از دست نداده است (از نظر FN وضعیت عالی).")
    else:
        print("- نتیجه: این مدل تعدادی بیمار پرخطر را از دست داده است؛ برای کاربرد بالینی باید با احتیاط بررسی شود.")
```

## 10) تعریف Cross-Validation و معیار بهینه‌سازی (F2)

در GridSearch:
- از StratifiedKFold استفاده می‌کنیم.
- معیار اصلی انتخاب بهترین مدل: **F2-score** با کلاس مثبت 1.


```python
scorer_f2 = make_scorer(fbeta_score, beta=2, pos_label=1)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)
```

## 11) مدل اول: SVM (SVC) با تست کرنل‌های مختلف

چرا SVM؟
- روی دیتاست‌های کوچک/متوسط با ویژگی‌های عددی معمولاً عملکرد قوی دارد.
- با Scaling بهتر می‌شود.

چه چیزی را تست می‌کنیم؟
- kernel: linear و rbf
- C: کنترل Regularization
- gamma: برای rbf مهم است
- class_weight: اگر به Recall کلاس 1 اهمیت بدهیم، balanced می‌تواند FN را کاهش دهد

هدف GridSearch:
- انتخاب بهترین تنظیمات با **F2-score** روی CV


```python
svm_pipe = Pipeline([
    ("prep", preprocess_all_numeric),
    ("clf", SVC())
])

svm_param_grid = {
    "clf__kernel": ["rbf", "linear"],
    "clf__C": [0.1, 0.5, 1, 2, 5, 10],
    "clf__gamma": ["scale", "auto"],
    "clf__class_weight": [None, "balanced"]
}

svm_gs = GridSearchCV(
    svm_pipe,
    svm_param_grid,
    scoring=scorer_f2,
    cv=cv,
    n_jobs=1
)

svm_gs.fit(X_train, y_train)

print("Best SVM params:", svm_gs.best_params_)
print("Best CV F2:", svm_gs.best_score_)

best_svm = svm_gs.best_estimator_
```

## 12) ارزیابی بهترین SVM روی داده تست

در این مرحله:
- خروجی نهایی SVM را با معیارهای پروژه گزارش می‌کنیم.
- Confusion Matrix را تحلیل می‌کنیم.


```python
evaluate_model(best_svm, X_test, y_test)
```

## 13) مدل دوم: KNN

چرا KNN؟
- ساده، قابل تفسیر، و روی برخی دیتاست‌ها Recall خوبی می‌دهد.
- به شدت به Scaling وابسته است (که انجام دادیم).

چه چیزی را تست می‌کنیم؟
- n_neighbors
- weights: uniform یا distance
- p: فاصله منهتن (1) یا اقلیدسی (2)

هدف:
- بیشینه کردن F2 (و کاهش FN)


```python
knn_pipe = Pipeline([
    ("prep", preprocess_all_numeric),
    ("clf", KNeighborsClassifier())
])

knn_param_grid = {
    "clf__n_neighbors": [3, 5, 7, 9, 11, 13],
    "clf__weights": ["uniform", "distance"],
    "clf__p": [1, 2]
}

knn_gs = GridSearchCV(
    knn_pipe,
    knn_param_grid,
    scoring=scorer_f2,
    cv=cv,
    n_jobs=1
)

knn_gs.fit(X_train, y_train)

print("Best KNN params:", knn_gs.best_params_)
print("Best CV F2:", knn_gs.best_score_)

best_knn = knn_gs.best_estimator_
```

## 14) ارزیابی بهترین KNN روی داده تست

اینجا بررسی می‌کنیم:
- آیا FN کم شده؟
- آیا Accuracy حداقل 85% هست؟
- آیا مجموع FP+FN با شرط پروژه سازگار است؟


```python
evaluate_model(best_knn, X_test, y_test)
```

## 15) مدل سوم: Decision Tree

چرا درخت تصمیم؟
- قابل تفسیرتر از SVM و KNN
- اما ممکن است overfit کند، پس باید پارامترهای کنترل پیچیدگی را تنظیم کنیم.

چه چیزی را تست می‌کنیم؟
- max_depth
- min_samples_split
- min_samples_leaf
- class_weight (برای کاهش FN ممکن است مفید باشد)

هدف:
- بهترین تنظیمات با معیار F2


```python
dt_pipe = Pipeline([
    ("prep", preprocess_all_numeric),
    ("clf", DecisionTreeClassifier(random_state=RANDOM_SEED))
])

dt_param_grid = {
    "clf__max_depth": [2, 3, 4, 5, 6, None],
    "clf__min_samples_split": [2, 5, 10, 20],
    "clf__min_samples_leaf": [1, 2, 4, 6],
    "clf__class_weight": [None, "balanced"]
}

dt_gs = GridSearchCV(
    dt_pipe,
    dt_param_grid,
    scoring=scorer_f2,
    cv=cv,
    n_jobs=1
)

dt_gs.fit(X_train, y_train)

print("Best DT params:", dt_gs.best_params_)
print("Best CV F2:", dt_gs.best_score_)

best_dt = dt_gs.best_estimator_
```

## 16) ارزیابی بهترین Decision Tree روی داده تست

درخت تصمیم ممکن است نسبت به SVM/KNN:
- Accuracy پایین‌تر یا نوسان بیشتر داشته باشد.
اما اگر FN را کم کند و F2 بالا بدهد، از نظر پزشکی ممکن است ارزشمند باشد.

با این حال شرط پروژه:
- Accuracy حداقل ~85%
- Falseها (FP+FN) کم (حداکثر حدود 15 وقتی Trueها حدود 40-50 هستند)


```python
evaluate_model(best_dt, X_test, y_test)
```

## 17) مقایسه نهایی مدل‌ها (تمرکز روی F2 و FN)

در این جدول:
- Accuracy و F2 و مقادیر TN/FP/FN/TP را کنار هم می‌گذاریم.
- برای پروژه‌ی سلامت، **FN و F2** معیار تصمیم‌گیری اصلی هستند.


```python
def extract_metrics(model, X_test, y_test):
    pred = model.predict(X_test)
    cm = confusion_matrix(y_test, pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    return {
        "Accuracy": accuracy_score(y_test, pred),
        "F2": fbeta_score(y_test, pred, beta=2, pos_label=1),
        "FN": fn,
        "FP": fp,
        "TN": tn,
        "TP": tp,
        "False_Total(FP+FN)": fp + fn
    }

results = pd.DataFrame({
    "SVM": extract_metrics(best_svm, X_test, y_test),
    "KNN": extract_metrics(best_knn, X_test, y_test),
    "DecisionTree": extract_metrics(best_dt, X_test, y_test)
}).T

display(results.sort_values(by=["F2", "Accuracy"], ascending=False))
```

## 18) تست چند حالت Train/Test Split (تحلیل حساسیت)

طبق صورت پروژه، لازم است مدل‌ها را با چند تقسیم متفاوت train/test بررسی کنیم.
در این بخش:
- چند `random_state` مختلف را امتحان می‌کنیم.
- برای هر مدل، با **همان بهترین هایپرپارامترهای انتخاب‌شده** دوباره fit می‌کنیم.
- روی تست: Accuracy، F2 و FN را گزارش می‌کنیم.

هدف در پروژه سلامت:
- **FN پایین** و **F2 بالا** حتی با تغییر تقسیم داده (پایداری مدل).


```python
from sklearn.base import clone

def metrics_quick(model, X_test, y_test):
    pred = model.predict(X_test)
    cm = confusion_matrix(y_test, pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    return {
        "Accuracy": accuracy_score(y_test, pred),
        "F2": fbeta_score(y_test, pred, beta=2, pos_label=1),
        "FN": fn,
        "FP": fp,
        "TN": tn,
        "TP": tp,
        "FP+FN": fp + fn
    }

models = {
    "SVM": best_svm,
    "KNN": best_knn,
    "DecisionTree": best_dt
}

random_states = [1, 7, 19, 42, 99]

rows = []
for rs in random_states:
    X_tr, X_te, y_tr, y_te = train_test_split(
        X, y, test_size=0.2, random_state=rs, stratify=y
    )
    for name, m in models.items():
        m2 = clone(m)              # همان مدل با همان تنظیمات
        m2.fit(X_tr, y_tr)
        met = metrics_quick(m2, X_te, y_te)
        rows.append({
            "random_state": rs,
            "model": name,
            **met
        })

split_results = pd.DataFrame(rows)
display(split_results.sort_values(by=["F2", "Accuracy"], ascending=False))

print("\nخلاصه میانگین‌ها روی random_state های مختلف:")
display(
    split_results.groupby("model")[["Accuracy", "F2", "FN", "FP+FN"]]
    .mean()
    .sort_values(by=["F2", "Accuracy"], ascending=False)
)
```

## نتیجه‌گیری نهایی پروژه

در این پروژه سه مدل طبقه‌بندی مختلف (SVM، KNN و Decision Tree) برای پیش‌بینی
شانس حمله قلبی بررسی شدند. معیار اصلی ارزیابی، با توجه به ماهیت پزشکی مسئله،
**F2-score** و به‌خصوص **تعداد False Negative (FN)** در نظر گرفته شد؛
زیرا FN به معنای تشخیص اشتباه یک بیمار پرخطر به‌عنوان کم‌خطر است
که می‌تواند پیامدهای جدی برای سلامت بیمار داشته باشد.

### مقایسه عددی مدل‌ها (روی داده تست)
- **SVM**
  - Accuracy ≈ 0.92
